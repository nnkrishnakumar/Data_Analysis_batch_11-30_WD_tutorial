```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
year=[2003,2004,2005,2006,2007,2008,2009,2010,2011,2012]
fee=[200,220,310,420,500,520,610,3001,4001,5500]
```

```python
plt.plot(year,fee)          # when we require line gharph then we use "plot"
plt.xlabel("Year",size=20, c='red')
plt.ylabel('fee',size=20,c='blue')
plt.xticks(ticks=[2003,2004,2005,2006,2007,2008,2009,2010,2011,2012],c='blue',size=10)
plt.yticks(c='red')
plt.show()
```

```python
plt.plot(year,fee,c='red', linewidth=5,linestyle=':')          # when we require line gharph then we use "plot"
plt.xlabel("Year",size=20, c='r')
plt.ylabel('fee',size=20,c='k')
plt.xticks(ticks=[2003,2004,2005,2006,2007,2008,2009,2010,2011,2012],c='blue',size=10)
plt.yticks(c='red')
plt.title("year-fee",size=20,c='purple')
plt.text(2004,5000,"SSVM",c='blue',size=20)
plt.show()
```

```python
# plt.figure(figsize=(15,12))
plt.plot(year,fee,c='red', linewidth=1,linestyle='-',marker='D',markerfacecolor='blue',markersize=5)          # when we require line gharph then we use "plot"
plt.xlabel("Year",size=20, c='r')
plt.ylabel('fee',size=20,c='k')
plt.xticks(ticks=[2003,2004,2005,2006,2007,2008,2009,2010,2011,2012],c='blue',size=10)
plt.yticks(c='red')
plt.title("year-fee",size=20,c='purple')
for i,j in zip(year,fee):
    plt.text(i,j+200,f"{j}",c='blue',size=10)
plt.show()
```
